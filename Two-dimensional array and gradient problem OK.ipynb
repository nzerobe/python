```python
import numpy as np
import matplotlib.pyplot as plt
```

# [Problem 1] Linear function

```python
x = np.arange(-50,50.1,step=0.1)
x = x.reshape((1001,1))
y = x/2+1
print(x.shape)
print(y.shape)
x
```

# [Problem 2] Array combination

```python
z = np.concatenate((x,y),axis=1)
z.shape
```

# [Problem 3] Find the gradient

```python
# used the array concatenated both x and y
gradient = (z[1:,1]-z[:-1,1])/(z[1:,0]-z[:-1,0])
gradient.shape
```

# [Problem 4] Draw a graph

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(10,5))
ax1.plot(z[:,0],z[:,1],color='green',linestyle='solid',linewidth=1)
ax1.set_title('Linear function')
ax1.set_xlim([-51, 51])
ax1.set_ylim([-25,27])
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax2.plot(z[1:,0],gradient,color='red',linestyle='dashed',linewidth=1)
ax2.set_title('Gradient')
ax2.set_xlabel("x")
ax2.set_ylabel("gradient")
plt.show()
```

# [Problem 5] Python functionalization

```python
def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    x = np.arange(*x_range)
    x = x.reshape((len(x),1))
    y = function(x)
    array_xy = np.concatenate((x,y),axis=1)
    gradient = (y[1:]-y[:-1])/(x[1:]-x[:-1])
    return array_xy, gradient
def function1(array_x):
    array_y = array_x**2
    return array_y
def function2(array_x):
    array_y = 2*array_x**2+2**array_x
    return array_y
def function3(array_x):
    array_y = np.sin(array_x**0.5)
    return array_y
array_xy1, gradient1 = compute_gradient(function1)
array_xy2, gradient2 = compute_gradient(function2)
array_xy3, gradient3 = compute_gradient(function3,(0,50.1,0.1))
print("f1 xy shape:",array_xy1.shape,"; gradient shape",gradient1.shape)
print("f2 xy shape:",array_xy2.shape,"; gradient shape",gradient2.shape)
print("f3 xy shape:",array_xy3.shape,"; gradient shape",gradient3.shape)
```

```python
fig, ((ax1,ax2),(ax3,ax4),(ax5,ax6)) = plt.subplots(3,2,figsize=(10,15))
ax1.plot(array_xy1[:,0],array_xy1[:,1],color='green',linestyle='solid',linewidth=1)
ax2.plot(array_xy1[1:,0],gradient1,color='red',linestyle='dashed',linewidth=1)
ax3.plot(array_xy2[:,0],array_xy2[:,1],color='green',linestyle='solid',linewidth=1)
ax4.plot(array_xy2[1:,0],gradient2,color='red',linestyle='dashed',linewidth=1)
ax5.plot(array_xy3[:,0],array_xy3[:,1],color='green',linestyle='solid',linewidth=1)
ax6.plot(array_xy3[1:,0],gradient3,color='red',linestyle='dashed',linewidth=1)
plt.show()
```

# [Problem 6] Find the minimum value

```python
print("min y1:",array_xy1[:,1].min(),"index:",array_xy1[:,1].argmin(),"min gradient:",gradient1.min(),"index:",gradient1.argmin())
print("min y2:",array_xy2[:,1].min(),"index",array_xy2[:,1].argmin(),"min gradient:",gradient2.min(),"index:",gradient2.argmin())
print("min y3:",array_xy3[:,1].min(),"index",array_xy3[:,1].argmin(),"min gradient:",gradient3.min(),"index:",gradient3.argmin())
```
