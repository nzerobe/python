# Problem 1

print total number of grains of rice received at 100th day

```python
import numpy as np
arr =[1]
for i in range(1,100):
    arr.append((arr[i-1])*2)
arr=np.array(arr)
n=np.sum(arr)
print("The total number of rice = {}".format(n))
```

```python
# Import libraries
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Configuring styles
sns.set_style("darkgrid")
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (18, 8)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
sum_arr=[]
for i in range(100):
    sum_arr.append(np.sum(arr[:i+1]))
```

```python
# Sample data
days = range(1, 101)

# First line
plt.plot(days, arr, 'b-x', linewidth=4, markersize=12, markeredgewidth=4, markeredgecolor='navy')

# Second line
plt.plot(days, sum_arr, 'r--o', linewidth=4, markersize=12,);

# Title
plt.title('Number of Rice')

# Line labels
plt.legend(['Number of rice you receive on that day', 'The total number of rice you receive by that day'])

# Axis labels
plt.xlabel('Days'); plt.ylabel('Number of Rice');
```

```python
#Let's take 200 the number of grains of rice required per one people to live one day
def number_of_days_to_live(number_of_grains,number_of_people):
    
    #Let's calculate the number of days to live for one people for the given number of grains
    days_to_live_per_one_people = number_of_grains/200
    
    #Let's calculate the number of days to live for the given number of people
    days_to_live = days_to_live_per_one_people/number_of_people
    
    return days_to_live 
```
