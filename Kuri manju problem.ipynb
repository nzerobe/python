Let assume that the solar system size is about 1000000 AU
    
    

```python
def solar_system(object_size):
    solar_size=1000000 # meter cube
    time=0
    while True:
        if object_size>=solar_size:
            break
        else:
            object_size*=2
            time+=5
    return print("{} Minutes".format(time))
```

```python
solar_system(300)
```

```python
def Tokyo_Dome(object_size):
    tokyo_dome_size=1240000 #meter cube
    time=0
    minutes=[]
    size=[]
    while True:
        if object_size>=tokyo_dome_size:
            break
        else:
            object_size*=2
            time+=5
            minutes.append(time)
            size.append(object_size)
    
    # Import libraries
    import matplotlib
    import matplotlib.pyplot as plt
    import seaborn as sns
    
    # Configuring styles
    sns.set_style("darkgrid")
    matplotlib.rcParams['font.size'] = 14
    matplotlib.rcParams['figure.figsize'] = (18, 8)
    matplotlib.rcParams['figure.facecolor'] = '#00000000'
    
    # Sample data
    
    # First line
    plt.plot(minutes, size, 'b-x', linewidth=4, markersize=12, markeredgewidth=4, markeredgecolor='navy')

    # Axis labels
    plt.xlabel('MINUTES'); plt.ylabel('SIZE OF OBJECT');
    plt.title("Tokyo Dome Situation")
    
    return print("{} Minutes".format(time))
```

```python
Tokyo_Dome(10000)
```
