[Problem 1]: Creating random numbers

```python
import numpy as np
mean = [-3, 0]
cov = [[1.0, 0.8],[ 0.8 ,1.0]]
```

```python
np.random.seed(0)
x1, x2 = np.random.multivariate_normal(mean,cov, 500).T
#plt.plot(x1, y2, '.')
#plt.show()
```

[Problem 2]: Visualization by scatter plot

```python
import matplotlib.pyplot as plt
plt.title("scatter")
plt.xlabel('x1')
plt.ylabel('x2')
plt.scatter(x1,x2)
plt.show()
```

[Problem 3]: Visualization by histogram

```python
plt.title("histogram of x1")
plt.xlim(-5,2) 
plt.xlabel('x1')
plt.ylabel('frequency')
plt.hist(x1)
plt.show()
```

```python
plt.title("histogram of x2")
plt.xlim(-5,2) 
plt.xlabel('x2')
plt.ylabel('frequency')
plt.hist(x2)
plt.show()
```

[Problem 4]: Addition of data

```python
mean = [0, -3]
cov = [[1.0, 0.8],[ 0.8 ,1.0]]
np.random.seed(0)
x3, x4 = np.random.multivariate_normal(mean,cov, 500).T
plt.scatter(x1,x2)
plt.scatter(x3,x4)
plt.legend(['0','1'])
plt.show()
```

[Problem 5]: Data combination

```python
x=np.concatenate((x1,x2), axis=0)
y=np.concatenate((x1,x2), axis=0)
x=x.reshape(1000,1)
y=y.reshape(1000,1)
data=np.concatenate((x,y), axis=1)
data.shape
```
