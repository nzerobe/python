# Calculating Thickness of the paper after folding 43times

```python
Thick=0.00008

Thick*=2**43
print("Folded_paper_Thickness in meters ={}m".format(Thick))
```

# Unit Conversion

```python
Thick=0.00008

Thick*=2**43
print("Folded_paper_Thickness in meters ={:.2f}Km".format(Thick/1000))
```

# Using For statement

```python
thick=0.00008
for i in range(43):
    thick=thick*2
print("Folded_paper_Thickness in Kilometers ={} Km".format(thick/1000))
```

# Comparison of Calculation times

```python
%%timeit
thick=0.00008
for i in range(43):
    thick=thick*2
```

```python
%%timeit
thick=0.00008
thick=thick*2**43
```

the best method is the second one, because it is faster.

# Visualization of process

# thick=0.00008
for i in range(1,44):
    print('Folding',i, '.',+thick*2**i)

# Save to list

```python
thick=0.00008
list =[]
for fold in range (1,45):
    list.append(thick*2**fold)
len(list)
```

# Display of line graph

```python
import  matplotlib.pyplot  as plt
%matplotlib inline
plt.title( " thick of folded paper " )
plt.xlabel( " number of folds " )
plt. ylabel( " thick [m] " )
plt.plot(list) # put list variable name in place of "list name" 
plt.show()
```

# Graph customization

```python
import matplotlib
import seaborn as sns
sns.set_style("darkgrid")
matplotlib.rcParams['font.size']=14
matplotlib.rcParams['figure.figsize'] = (9,5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```
